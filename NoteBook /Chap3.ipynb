```python
#Exo1
L = [1, 2]
L3 = 3 * L  # Répète la liste trois fois

print("L3 =", L3)        # ➤ [1, 2, 1, 2, 1, 2]
print("L3[0] =", L3[0])  # ➤ 1
print("L3[-1] =", L3[-1])  # ➤ 2
# print("L3[10]")        # ➤ Erreur: IndexError (car L3 a seulement 6 éléments)

# Création d’une nouvelle liste avec les carrés de chaque élément
L4 = [k**2 for k in L3]  # ➤ [1, 4, 1, 4, 1, 4]

# Concaténation des deux listes
L5 = L3 + L4
print("L5 =", L5)        # ➤ [1, 2, 1, 2, 1, 2, 1, 4, 1, 4, 1, 4]
```

```python
#Ex04
m = 5
L = [n - m/2 for n in range(m)]  # ➤ [-2.5, -1.5, -0.5, 0.5, 1.5]
ans = 1 + L[0] + L[-1]           # ➤ 1 + (-2.5) + 1.5 = 0
print("ans =", ans)  # ➤ 0.0
```

```python
#Ex0 5
import numpy as np
import matplotlib.pyplot as plt

h = 0.001
a = -0.5
N = 1000

# Initial values: e^0, e^ha, e^2ha
u = [np.exp(0), np.exp(h * a), np.exp(2 * h * a)]

# Calcul par récurrence
for n in range(N):
    unp3 = u[-1] + h * ((23 / 12) * u[-1] - (4 / 3) * u[-2] + (5 / 12) * u[-3])
    u.append(unp3)

t = np.array([n * h for n in range(N + 3)])
exact = np.exp(a * t)
error = np.abs(exact - u)

# Tracé des résultats
plt.plot(t, u, label='Approximation')
plt.plot(t, exact, label='Exact')
plt.title('Solution numérique vs exacte')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.grid()
plt.legend()
plt.show()

# Tracé de l'erreur
plt.plot(t, error, label='Erreur absolue')
plt.title('Erreur |e^(-at) - u_n|')
plt.xlabel('t')
plt.ylabel('Erreur')
plt.grid()
plt.legend()
plt.show()
```

```python
#Exo7
def symmetric_diff(A, B):
    return (A - B) | (B - A)

A = {1, 2, 3}
B = {2, 3, 4}

print("Résultat perso :", symmetric_diff(A, B))          # ➤ {1, 4}
print("Résultat intégré :", A.symmetric_difference(B))   # ➤ {1, 4}
```

```python
#Exo8 
S = {1, 2, 3}
T = {2, 3, 4}

print("intersection =", S.intersection(T))  # ➤ {2, 3}

# Intersection in-place
S.intersection_update(T)
print("S modifié =", S)                     # ➤ {2, 3}
```
